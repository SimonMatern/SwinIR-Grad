```python
from models.model import DeepAggNet
from models.network_swinir import SwinIR3D
from einops import rearrange
import torch 
import glob
from evaluation import TestSet, model_from_ckpt, evaluation
import torch.nn.functional as F
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import util_calculate_psnr_ssim as util
from tqdm import tqdm
from collections import OrderedDict

import os 
data_dirs = ["/shared/DIV2K/TestSet/BSDS100/","/shared/DIV2K/TestSet/Set14/", "/shared/DIV2K/TestSet/Set5/", "/shared/DIV2K/TestSet/urban100/", "/shared/DIV2K/TestSet/manga109//"]

ckpts = glob.glob("lightning_logs/version*")
for N in [2, 4, 8, 10, 16, 20, 30, 40, 50]:
    for ckpt_path in ckpts:
        torch.cuda.empty_cache()
        if os.path.isfile(f'{ckpt_path}/{N}.csv'):
            print("Evaluation already exists")
            continue
        try:
            evaluation(ckpt_path, data_dirs, N=N,  samples=1, device="cuda")
        except RuntimeError as e:
            if 'out of memory' in str(e):
                print('| WARNING: ran out of memory, retrying on CPU')
                evaluation(ckpt_path, data_dirs, N=N,  samples=1, device="cpu")

        except Exception as e:
            print('An exception occurred in {}: {}'.format(ckpt_path,e))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import yaml
import re
ckpts = glob.glob("lightning_logs/version*")

results = {}
for ckpt in ckpts:
    with open(ckpt+ "/config.yaml", 'r') as stream:
        config = yaml.safe_load(stream)

    csvs = glob.glob(ckpt+"/[0-9]*.csv")
    if len(csvs)==0:
        continue
    
    Ns = [re.findall(r'\d+', csv)[-1] for csv in csvs]
    dfs = [pd.read_csv(csv) for csv in csvs]
    dfs = [dfs[i].assign(**{"N":Ns[i]}) for i in range(len(dfs))]
    df = pd.concat(dfs)
    
    result = {"df":df, "config":config }
    augm = config["data"]["augm"]
    results.setdefault(augm, []).append(result)

metrics = ["psnr","ssim","psnr_y","ssim_y","psnrb","psnrb_y"]
for augm in results:
    for metric in metrics:
        for result in results[augm]:
            df = result["df"]
            config = result["config"]
            
            mean = df.groupby('N')[metric].agg(['mean', 'std'])
            mean.index = mean.index.astype(int)
            mean = mean.sort_index()
            print(mean)
            plt.plot(mean.index.values, mean["mean"], label =config["trainer"]["name"])
        plt.title(aug+":"+metric)
        plt.legend()
        plt.show()
```
